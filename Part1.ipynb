```python
#Sierpinski triangle
import numpy as np
import matplotlib.pyplot as plt

#initialize N
N =10000

#initialize vectors sx and sy
sx = np.zeros(N)
sy = np.zeros(N)

for i in range(1,N):
  #generate random integer 
  k = np.random.randint(1,4)
  
  sx[i] = sx[i-1]/2+ k-1
  sy[i] = sy[i-1]/2
  if k==2:
    sy[i]+=2
plt.plot(sx,sy,'k.', markersize= 1)
plt.title('Sierpinski Triangle')
plt.axis('off')
plt.show()
```
